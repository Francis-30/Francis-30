```python
import pandas as pd
import plotly.express as px
```

```python
path = '/content/drive/MyDrive/Churn_Modelling.csv'
```

```python
churn_df = pd.read_csv(path)
churn_df.head(10)
```

```python
churn_df.tail(10)
```

```python
churn_df.info()
```

```python
churn_df.describe()
```

```python
churn_df.drop('RowNumber',inplace=True,axis=1)
```

```python
churn_df.Geography.unique()
```

Exited means customer no longer use the bank
**bold text**

```python
data = churn_df.groupby(['Geography','Exited'])['Exited'].count().reset_index(name = 'count')
data
```

```python
fig = px.bar(data, x='Geography', y='count', color='Exited')
fig.show()
```

```python
churn_df.Exited.unique()
```

```python
churn_df['status']=churn_df.Exited.replace(1,'Exited').replace(0,'Retained')
```

```python
churn_df.head(10)
```

```python
churn_df.status.unique()
```

```python
data = churn_df.groupby(['Geography','status'])['status'].count().reset_index(name='count')
data
```

```python
fig = px.bar(data, x='Geography', y='count', color='status')
fig.show()
```

```python
data= churn_df.groupby(['Geography'])['status'].value_counts(normalize= True).reset_index(name='percentage')
data
```

```python
fig = px.bar(data, x='Geography', y ='percentage', color = 'status')
fig.show()
```

```python
data = churn_df.groupby('status')['status'].count().reset_index(name='count')
data
```

```python
data = churn_df.groupby('status')['status'].count().reset_index(name='count')
data
```

```python
fig = px.pie(data, values='count', labels='status')
fig.show()
```
